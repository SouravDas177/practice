```python
from sklearn.datasets import make_classification
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
x,y=make_classification(n_samples=1000,n_features=5,n_informative=3,n_redundant=2,random_state=1)
```

```python
x
```

```python
df=pd.DataFrame(x)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x=df[1],y=df[2])
```

```python
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
model=DecisionTreeClassifier(max_depth=7,max_leaf_nodes=8,criterion="gini",splitter="best")
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

print("accuracy is -->",accuracy_score(y_test,y_pred))
print("confusion matrix is -->",confusion_matrix(y_test,y_pred))
print("precision score is -->",precision_score(y_test,y_pred))
print("recall score is -->",recall_score(y_test,y_pred))
```

```python
print((cross_val_score(model,x_train,y_train,cv=5)).mean())
```
