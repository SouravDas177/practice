```python
import numpy as np
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/placement-dataset.csv')
```

```python
df=df.dropna()
```

```python
sns.scatterplot(x=df["cgpa"],y=df["iq"],hue=df["placement"],size=df["placement"])
plt.xlabel("cgpa", rotation=45)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop="first")

part=ohe.fit_transform(df[["city"]]).toarray()
part_df=pd.DataFrame(part,columns=ohe.get_feature_names_out())
print(part_df)
```

```python
df=pd.concat([df,part_df],axis=1)
```

```python
df.head()
```

```python
df=df.drop("city",axis=1)
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split

x=df.drop("placement",axis=1)
y=df["placement"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pres=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score

print("accuracy is --> ",accuracy_score(y_test,y_pres))
print("confusion matrix is --> ",confusion_matrix(y_test,y_pres))
print("precision is --> ",precision_score(y_test,y_pres))
print("recall is --> ",recall_score(y_test,y_pres))
print("f1 score is --> ",f1_score(y_test,y_pres))

from sklearn.model_selection import cross_val_score

print("cv is --> ",(cross_val_score(model,x_train,y_train,cv=5)).mean())

```
