```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
dic={
    "gender":["f","f","m","f","m","m"],
    "ocupation":["student","programers","programers","programers","student","student"],
    "suggition":["pubg","git","wp","git","pubg","pubg"]
}
```

```python
df=pd.DataFrame(dic)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
label=LabelEncoder()
df["gender"]=label.fit_transform(df["gender"])
df
```

```python
df["ocupation"]=df["ocupation"].map({"student":0,"programers":1})
```

```python
df["suggition"]=df["suggition"].map({"pubg":0,"git":1,"wp":2})
```

```python
model=DecisionTreeClassifier()
model.fit(df[["suggition","ocupation"]],df["gender"])
```

```python
df.head()
```

```python
plot_tree(model)
```
